```python
import os
import json
import pandas as pd
import pymysql
```

```python
#Aggregated_Insurance
#Getting the data
path1 = "C:/Users/hp/pulse/data/aggregated/insurance/country/india/state/"
agg_ins_list= os.listdir(path1)

columns1 = {"States":[], "Years":[],"Quarter":[], "Transaction_type":[],
             "Transaction_count":[], "Transaction_amount":[]
             }
#Extracting state folder data
for state in agg_ins_list:
    cur_states = path1 + state + "/"
    agg_year_list = os.listdir(cur_states)

    #Extracting year folder data
    for year in agg_year_list:
        cur_year= cur_states + year + "/"
        agg_file_list= os.listdir(cur_year)

        #Extracting json file
        for file in agg_file_list:
            cur_file = cur_year + file
            data= open(cur_file,"r")

            A = json.load(data)

            #GETTING THE COLUMNS OF JSON FILE
            for i in A["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                columns1["Transaction_type"].append(name)
                columns1["Transaction_count"].append(count)
                columns1["Transaction_amount"].append(amount)
                columns1["States"].append(state)
                columns1["Years"].append(year)
                columns1["Quarter"].append(int(file.strip(".json")))
            
#Creating the dataframe
Agg_insurance= pd.DataFrame(columns1)

#Cleaning data(removing - from state and title capitalization )
Agg_insurance["States"] = Agg_insurance["States"].str.replace('andaman-&-nicobar-islands', 'Andaman & Nicobar Islands')
Agg_insurance["States"] = Agg_insurance["States"].str.replace("-"," ")
Agg_insurance["States"] = Agg_insurance["States"].str.title()
Agg_insurance["States"] = Agg_insurance["States"].str.replace('Dadra & Nagar Haveli & Daman & Diu', 'Dadra and Nagar Haveli and Daman and Diu')
        
```

```python
# Path to the aggregated user data
path3 = "C:/Users/hp/pulse/data/aggregated/user/country/india/state/"
agg_user_list = os.listdir(path3)

# Initialize columns for the DataFrame
columns3 = {
    "States": [], "Years": [], "Quarter": [],  "Brands": [], "Transaction_count": [],  
    "RegisteredUsers": [], "AppOpens": [], "Percentage": []
}

# Extracting state folder data
for state in agg_user_list:
    cur_states = os.path.join(path3, state)
    agg_year_list = os.listdir(cur_states)

    # Extracting year folder data
    for year in agg_year_list:
        cur_year = os.path.join(cur_states, year)
        agg_file_list = os.listdir(cur_year)

        # Extracting JSON file data
        for file in agg_file_list:
            cur_file = os.path.join(cur_year, file)
            
            # Open the file and load JSON data
            with open(cur_file, "r") as data:
                C = json.load(data)

            # Extract aggregated data
            aggregated_data = C["data"]["aggregated"]
            registered_users = aggregated_data["registeredUsers"]
            app_opens = aggregated_data["appOpens"]

            # Check if 'usersByDevice' exists and is not None
            device_data = C["data"].get("usersByDevice", None)
            
            if device_data is not None:  # Proceed only if device_data is a valid list
                # Append data to columns3 for each device
                for device in device_data:
                    columns3["Brands"].append(device["brand"])
                    columns3["Transaction_count"].append(device["count"])  
                    columns3["Percentage"].append(device["percentage"])
                    columns3["RegisteredUsers"].append(registered_users)
                    columns3["AppOpens"].append(app_opens)
                    columns3["States"].append(state)
                    columns3["Years"].append(year)
                    columns3["Quarter"].append(int(file.strip(".json")))
            else:
                print(f"Missing or invalid 'usersByDevice' in file: {cur_file}")

# Create the DataFrame from the columns
Agg_user = pd.DataFrame(columns3)

#Cleaning data(removing - from state and title capitalization )
Agg_user["States"] = Agg_user["States"].str.replace('andaman-&-nicobar-islands', 'Andaman & Nicobar Islands')
Agg_user["States"] = Agg_user["States"].str.replace("-"," ")
Agg_user["States"] = Agg_user["States"].str.title()
Agg_user["States"] = Agg_user["States"].str.replace('Dadra & Nagar Haveli & Daman & Diu', 'Dadra and Nagar Haveli and Daman and Diu')


```

```python
#MAP_INSURANCE
path4= "C:/Users/hp/pulse/data/map/insurance/hover/country/india/state/"
map_ins_list = os.listdir(path4)

# Dictionary to store extracted transaction data
columns4 = {
    'States': [], 'Years': [], 'Quarter': [], "Districts": [],
    'Transaction_count': [], 'Transaction_amount': []
}

# Loop through each state
for state in map_ins_list:
    state_path = os.path.join(path4, state)
    map_year_list = os.listdir(state_path)

    # Loop through each year
    for year in map_year_list:
        year_path = os.path.join(state_path, year)
        map_file_list = os.listdir(year_path)

        # Loop through each quarter (JSON file)
        for file in map_file_list:
            file_path = os.path.join(year_path, file)
            with open(file_path, 'r') as f:
                D = json.load(f)

            quarter = int(file.strip('.json'))

            #extracting the hover transaction data
            for i in D["data"]["hoverDataList"]:
                name = i["name"]
                count = i['metric'][0]['count']
                amount = i['metric'][0]['amount']
                columns4['Districts'].append(name)
                columns4['Transaction_count'].append(count)
                columns4['Transaction_amount'].append(amount)
                columns4['States'].append(state)
                columns4['Years'].append(int(year))
                columns4['Quarter'].append(quarter)

Map_insurance = pd.DataFrame(columns4)
                
#Cleaning data(removing - from state and title capitalization )
Map_insurance["States"] = Map_insurance["States"].str.replace('andaman-&-nicobar-islands', 'Andaman & Nicobar Islands')
Map_insurance["States"] = Map_insurance["States"].str.replace("-"," ")
Map_insurance["States"] = Map_insurance["States"].str.title()
Map_insurance["States"] = Map_insurance["States"].str.replace('Dadra & Nagar Haveli & Daman & Diu', 'Dadra and Nagar Haveli and Daman and Diu')
```
